**Practical 5 : Develop a cost function of linear regression using sample dataset.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([3, 2, 3, 4, 6])

# Plot the data points
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Data')
plt.show()

# Number of data points
m = len(Y)

# Initial parameter values
theta0 = 0.0  # Intercept
theta1 = 0.0  # Slope

def hypothesis(x, theta0, theta1):
    """Compute the hypothesis for a given input x."""
    return theta0 + theta1 * x

def cost_function(theta0, theta1, X, Y):
    """Compute the Mean Squared Error cost for linear regression."""
    total_cost = 0
    for i in range(m):
        total_cost += (hypothesis(X[i], theta0, theta1) - Y[i]) ** 2
    return total_cost / (2 * m)

# Evaluate cost for a range of theta1 values while keeping theta0 fixed
theta1_values = np.linspace(-10, 12, 50)
cost_values = []
for t1 in theta1_values:
    cost = cost_function(theta0, t1, X, Y)
    cost_values.append(cost)

# Plot the cost function with respect to theta1
plt.plot(theta1_values, cost_values, label='Cost Function')
plt.xlabel('Theta1')
plt.ylabel('Cost')
plt.title('Cost Function for Linear Regression')
plt.legend()
plt.show()

# Calculate and print the initial cost (with theta0 = 0 and theta1 = 0)
initial_cost = cost_function(theta0, theta1, X, Y)
print(f"Initial Cost (MSE) with theta0 = {theta0} and theta1 = {theta1}: {initial_cost}")

# Evaluate the cost using experimental parameters
theta0_exp = 0.2
theta1_exp = 0.8
final_cost = cost_function(theta0_exp, theta1_exp, X, Y)
print(f"Final Cost (MSE) with theta0 = {theta0_exp} and theta1 = {theta1_exp}: {final_cost}")
```

**Practical 6 : Develop a Gradient descent of linear regression using sample dataset.**
